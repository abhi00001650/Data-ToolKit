```python
#Can u merge 2 Dataframe ?? >>> Yes
```

```python
#Merge
df1 = pd.DataFrame({'key1' : [1,2,4,5,6],
                    'key2' : [4,5,6,7,8],
                    'key3' : [3,4,5,6,6]
                   })
df1
```

```python
df2 = pd.DataFrame({'key1' : [1,2,45,6,67],
                    'key4' : [56,5,6,7,8],
                    'key5' : [3,56,5,6,6]
                   })
df2
```

```python
#INNER MERGE>>
pd.merge(df1,df2, how = 'inner')
```

```python
pd.merge(df1,df2, how = 'left') # left merge
```

```python
pd.merge(df1,df2, how = 'right') # right merge
```

```python
pd.merge(df1,df2, how = 'outer') # outer merge >> all the elements of the common column from both the data frame
```

```python
pd.merge(df1,df2, how = 'outer' , left_on= "key2" , right_on= "key4")
```

```python
#Join >> join works based on index | merge works based on columns
```

```python
#JOIN
df1 = pd.DataFrame({'key1' : [1,2,4,5,6],
                    'key2' : [4,5,6,7,8],
                    'key3' : [3,4,5,6,6]},
                    index = ['a','b','c','d','e']
                   )
df1
```

```python
df2 = pd.DataFrame({'key1' : [1,2,45,6,67],
                    'key4' : [56,5,6,7,8],
                    'key5' : [3,56,5,6,6]
                   }, index = ['a','b','h','i','j'])
df2
```

```python
df1.join(df2, how = "inner") # Syntax
```

```python
df1.join(df2, how = "left") # Syntax
```

```python
df1.join(df2, how = "right") # Syntax
```

```python
df1.join(df2, how = "outer") # Syntax
```

```python
df1.join(df2, how = "cross") # Syntax
```
