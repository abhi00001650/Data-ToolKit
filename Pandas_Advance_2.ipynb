```python
data1 = {
    'A' : [1,2,None,4,5,None,7,8,9,10],
    'B' : [None,11,12,13,None,15,16,None,18,19]
}
df2 = pd.DataFrame(data1)
```

```python
df2
```

```python
df2['A'].mean()
```

```python
#fill NaN with mean value
df2['A'].fillna(df2['A'].mean())
```

```python
#fill NaN with median
df2['A'].median()
```

```python
df2['A'].fillna(df2['A'].median())
```

```python
df2
```

```python
df2.fillna("Some value")
```

```python
df2.fillna(0)
```

```python
#Another way to fill is>> ffill( forwardfill ) >> start seeing from the bottom of the dataframe and fill with the next observed value
df2.fillna(method = "ffill")
```

```python
#Another way to fill is>> bfill( backwardfill ) >> see the data from top and fill with next observed value
df2.fillna(method = "bfill")
```

```python
#What if i ask max , min , std , age/fare , in one result
#Function name is Groupby >>
```

```python
import numpy as np
df.groupby(['Survived'])['Fare'].agg([min, 'max' , 'median' , 'count' , np.std , 'var'])
```

```python
#Groupby based on 2 columns
#Q How many passengers survived or not survived for each sex and passenger class
df.groupby(['Sex' , 'Pclass'])['Survived'].sum()
```

```python
#covert that into dataframe 
df.groupby(['Sex' , 'Pclass'])['Survived'].sum().to_frame()
```

```python
#to reset index
df.groupby(['Sex' , 'Pclass'])['Survived'].sum().to_frame().reset_index()
```
